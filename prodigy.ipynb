```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
file_path = "/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
data = pd.read_csv(file_path)

# ===== Example 1: Bar chart for categorical variable =====
# Replace 'Region' with any categorical column name
category_counts = data['Region'].value_counts().dropna()

plt.figure(figsize=(10,6))
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Distribution of Countries by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.show()

# ===== Example 2: Histogram for continuous variable =====
# Replace 'Population' with any continuous numeric column name
# (For your file, you may need a different dataset with numeric values like age or population)
# Example: histogram of some numeric column if available
if 'Population' in data.columns:
    plt.figure(figsize=(8,6))
    plt.hist(data['Population'].dropna(), bins=20, color='lightgreen', edgecolor='black')
    plt.title("Distribution of Population")
    plt.xlabel("Population")
    plt.ylabel("Frequency")
    plt.show()
```
